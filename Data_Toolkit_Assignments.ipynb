#Q2. How does broadcasting work in NumPy?
ANS: Broadcasting in NumPy allows arithmetic operations between arrays of different shapes by automatically expanding their dimensions to be compatible.

Rules:

1.**Align dimensions:** Start from the rightmost dimension.

2.**Compare sizes:** Dimensions must either be equal or one of them must be 1.

3.**Expand dimensions:** If one array has a size of 1 in a dimension, it is stretched to match the other array.

This enables efficient element-wise operations without explicitly reshaping arrays.

#Q6. What are the differences between NumPy arrays and Python lists?
ANS: NumPy arrays and Python lists differ in several key ways:

1.**Performance:**

 * NumPy arrays are faster and more efficient for numerical operations due to their fixed type and memory layout.

* Python lists are slower for numerical computations as they can hold different types and have more overhead.

2.**Functionality:**

* NumPy arrays support vectorized operations (element-wise operations) and advanced mathematical functions.

* Python lists require loops or list comprehensions for similar operations.

3.**Memory Usage:**

* NumPy arrays use less memory as they store homogeneous data types.

* Python lists use more memory due to their flexibility in storing heterogeneous data types.

4.**Size and Shape:**

* NumPy arrays have a fixed size and shape, enabling operations like reshaping and broadcasting.

* Python lists are dynamic in size and do not support these operations natively.

5.**Library Support:**

* NumPy arrays integrate well with scientific computing libraries (e.g., SciPy, Pandas).

* Python lists are more general-purpose and lack this integration.

#Q17.  Explain the difference between apply() and map() in Pandas.
ANS: Both apply() and map() are used to apply functions to Pandas data structures, but they have key differences in usage and scope.

**1️⃣ map() - Works on Series (Single Column)**

* Used only for Pandas Series (single column).

* Applies a function element-wise to each value.

* Can accept a function, dictionary, or a mapping (like a lambda function).

**2️⃣ apply() - Works on Both Series and DataFrames**

* Can be used on both Series and DataFrames.

* Works on rows (axis=1) or columns (axis=0) in a DataFrame.

* More powerful, as it can handle complex functions.

#Q21.Why is NumPy’s array slicing faster than Python’s list slicing?
ANS: NumPy array slicing is faster than Python list slicing due to how data is stored and handled in memory, and the optimizations that NumPy leverages. NumPy arrays store elements of the same data type in contiguous memory blocks, while Python lists store elements as pointers to objects that can be scattered in memory. This contiguous storage allows for more efficient data access and manipulation.

#Q4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [1, 1, 2, 2, 3]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()


plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')


plt.show()
```

#Q5.Generate a bar plot using Plotly.

```python
import plotly.express as px
import pandas as pd

data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 15, 25]
}

df = pd.DataFrame(data)


fig = px.bar(df, x='Category', y='Values', title='Bar Plot Example', text='Values')


fig.update_traces(textposition='outside')
fig.update_layout(xaxis_title='Categories', yaxis_title='Values', showlegend=False)


fig.show()
```

#Q8. Create a line plot with multiple lines using Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x + 0.5)

plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-')
plt.plot(x, y2, label='cos(x)', color='red', linestyle='--')
plt.plot(x, y3, label='sin(x + 0.5)', color='green', linestyle=':')


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Multiple Lines')

plt.legend()
plt.show()
```

#Q9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

```python
import pandas as pd


data = {
    'Name': ['Deepak', 'Raushan', 'Chandan', 'Vikash', 'Ankur'],
    'Age': [22, 25, 28, 32, 35],
    'Salary': [50000, 60000, 70000, 80000, 90000]
}

df = pd.DataFrame(data)


threshold = 70000


filtered_df = df[df['Salary'] > threshold]

print("Original DataFrame:")
print(df)
print("\nFiltered DataFrame (Salary > {}):".format(threshold))
print(filtered_df)
```
